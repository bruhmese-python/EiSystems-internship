```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("Bmi_male_female.csv")
df.head()
```

```python
x= df[["Height","Weight"]]
y= df.Gender
```

```python
x.head()
```

```python
y.head()
```

```python
Y = pd.get_dummies(y)
Y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
model = DecisionTreeClassifier()
model.fit(x,Y)
```

```python
model.predict([[190,60]])
```

```python
print("Female") if model.predict([[190,60]])[0][0] else print("Male")
```
