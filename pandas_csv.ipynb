Data Cleaning

```python
df = pd.read_excel("Temper.xlsx")
```

```python
df
```

```python
#you could use replace function for replacing with a constant - but hey this is a generic way
```

```python
median = df.Temperature.median()
df.Temperature = df.Temperature.apply(lambda x,median=median: median if (x - median > median) else x)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df1 = pd.read_excel("Temper.xlsx")
#df1.Temperature = df1.Temperature.apply(lambda x: np.nan if x == 33333 else x)
df1 = df1.replace([33333,11111],np.NaN)
```

```python
df1.dropna(thresh=3,ignore_index=True,inplace=True)
#inplace assigns the values. Avoids df1 = ...
#atleast 3 non missing values (row wise) - so in asingle row wont keep more than 1 missing values
```

Merge 2 or more dataframes together

```python
data1 = {'name': ['a', 'b', 'c', 'd', 'e', 'f'],
         'age': [4, 5, 16, 24, 15, 36],
         'sub': ['s1', 's2', 's3', 's4', 's5', 's6']}
d1 = pd.DataFrame(data1)

# Creating variations for data2
data2 = {'name': ['a', 'h', 'i', 'j', 'k'],
         'age': [7, 8, 9, 10, 11],
         'sub': ['s7', 's8', 's9', 's10', 's11']}
d2 = pd.DataFrame(data2)
```

```python
d1
```

```python
d2
```

```python
pd.concat([d1,d2],ignore_index=True,axis=0)
```

```python
pd.concat([d1,d2],ignore_index=True,axis=1)
```

```python
pd.concat([d1,d2],ignore_index=True,axis=0)
```

```python
pd.merge(d1,d2,on="name",how="left") 
```

```python
pd.merge(d1,d2,on="name",how="outer") 
```

```python
pd.merge(d1,d2,on="name",how="inner") 
```

```python
pd.merge(d1,d2,on="name",how="right") 
```
