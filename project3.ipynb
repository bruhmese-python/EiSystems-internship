```python
import pandas as pd
insurance = pd.read_csv("insurance.csv")
```

```python
insurance.head()
```

```python
insurance.isna()
```

```python
x=insurance[["age"]]
y=insurance[["bought_insurance"]]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y, color='white', edgecolor='orange')
plt.xlabel('age')
plt.ylabel('bought')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
model  = LogisticRegression()
model.fit(x,y)
```

```python
model.predict([[52]])
```
