```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
likes=[190,211,590,321,567,333,812]
```

```python
plt.title("My Facebook Likes")
plt.xlabel("Number of days")
plt.ylabel("Number of Likes")
plt.plot(days,likes,color="red",marker="o",mfc="blue",mec="yellow",mew=3,ms=15,ls="--",lw=5,fillstyle="full")
plt.grid(color="green")
plt.xlim(3,5)
plt.ylim(300,500)
```

comparing different social media using matplotlib

```python
days
```

```python
fb=likes.copy()
insta=[210,511,178,267,712,345,812]
ytube=[910,210,689,490,512,680,222]
```

```python
fb
insta
ytube
```

```python
plt.title("Social Media Analyzation")
plt.xlabel("Number of days")
plt.ylabel("Number of likes")
plt.plot(days,fb,label="fbook",color="blue",marker="o",ms=10,mfc="yellow")
plt.plot(days,insta,label="insta",color="m",marker="s",ms=10,mfc="pink")
plt.plot(days,ytube,label="ytube",color="red",marker=">",ms=10,mfc="black")
plt.legend(loc="upper right")
plt.grid(color="orange")
#plt.ylim(500,600)
#plt.xlim(3,5)
plt.savefig("mygraph.png")
plt.show()
```

Pie chart

```python
likes=[250,190,470,550,390]
p=["insta","fbook","sc","ln","yt"]
```

```python
plt.pie(likes,labels=p,colors=["pink","blue","yellow","cyan","red"],autopct="%1.1f",shadow=True,explode=[0.2,0.3,0,0,0],radius=0.5)
plt.show()
```
